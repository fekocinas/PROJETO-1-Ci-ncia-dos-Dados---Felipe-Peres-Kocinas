# Introdução:

### O que o IDH representa?

O IDH (Índice de Desenvolvimento Humano) é um dado calculado pelo Programa das Nações Unidas para o Desenvolvimento (PNUD), órgão da ONU,e é formado a partir de três fatores: expectativa de vida, renda per capita e anos médios de estudo e anos esperados de escolaridade. 

### Coleta dos dados:

Os dados foram coletados no website Gapminder e escolhidos para serem comparados ao IDH, tais foram: a porcentagem da população que vive em regiões urbanas e o total do PIB (Produto Interno Bruto) em dólares de cada país. 

A partir desses dados foi realizada a seguinte pergunta a ser respondida nesse documento. 
### Pergunta:
O IDH dos países é influênciado pelo índice populacional das cidades e se o PIB dos países for baixo é possível seu IDH ser elevado?


O motivo da escolha da pergunta foi porque poderia analisar situações diferentes da realidade brasileira, pelo fato de que o Brasil é um país majoritariamente urbano e possui um PIB muito alto.

A minha pergunta foi formulado a partir da hipótese de que países mais urbanizados,  passaram pelo processo de Urbanização, e isso possibilitou o aumento da geração de trabalhos nas cidades, o que impulsionou a economia, e o aumento da facilidade do acesso a escolas e hospitais, fatores que são fortemente ligados aos dados que compõem o IDH, e a segunda parte da pergunta foi formulada para saciar uma curiosidade sobre países com baixo PIB, para descobrir a tendência de seus IDHs.

### Código:

A seguir foram implementados os códigos necessários para que fosse possível abrir e juntar em jupyter com o uso da biblioteca pandas os dados escolhidos e que foram baixados do Gapminder. Também foram executados códigos para selecionar os dados apenas do ano de 2014 e atribuir títulos as colunas.

```python
# %matplotlib inline
#------Import das bibliotecas-----
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#------abrindo os arquivos do excel-------
IDH = pd.read_excel('hdi_human_development_index.xlsx')
PIB = pd.read_excel('total_gdp_us_inflation_adjusted.xlsx')
PopUrb = pd.read_excel('urban_population_percent_of_total (1).xlsx')
```

```python
#------selecinando os dados desejados de cada arquivo------
IDH_2014 = IDH.loc[:, ["geo", 2014]]
PIB_2014 = PIB.loc[:, ["geo", 2014]]
PopUrb_2014 = PopUrb.loc[:, ["geo", 2014]]
```

```python
#-------selecionando a coluna do index-------
IDH_idx = IDH_2014.set_index('geo')
PIB_idx = PIB_2014.set_index('geo')
PopUrb_idx = PopUrb_2014.set_index('geo')
```

```python
#------unindos os arquivos e mudando o nome das colunas-------
df = IDH_idx.join(PopUrb_idx, how='inner', lsuffix='_IDH', rsuffix='_% PopUrb')
df["IDH"]=df.pop("2014_IDH")
df["% da População Urbana"]=df.pop('2014_% PopUrb')

dados = df.join(PIB_idx, how='inner')
dados["PIB total em dólares"]=dados.pop(2014)

print("Todos os dados obtidos estão a seguir:")
# comando para tirar os NaN da tabela
dados1 = dados.dropna()
dados1
```

# Códigos para o ano de 2000:

```python
IDH_2000 = IDH.loc[:, ["geo", 2000]]
PIB_2000 = PIB.loc[:, ["geo", 2000]]
PopUrb_2000 = PopUrb.loc[:, ["geo", 2000]]

IDH_idx1 = IDH_2000.set_index('geo')
PIB_idx1 = PIB_2000.set_index('geo')
PopUrb_idx1 = PopUrb_2000.set_index('geo')

df1 = IDH_idx1.join(PopUrb_idx1, how='inner', lsuffix='_IDH', rsuffix='_% PopUrb')
df1["IDH"]=df1.pop("2000_IDH")
df1["% da População Urbana"]=df1.pop('2000_% PopUrb')

dados2 = df1.join(PIB_idx1, how='inner')
dados2["PIB total em dólares"]=dados2.pop(2000)

# comando para tirar os NaN da tabela
dados3 = dados2.dropna()
```

# Gráfico 1 -  A porcentagem da população urbana pelo IDH - 2014

```python
dados1.plot.scatter(x="IDH", y="% da População Urbana", grid=True, title="A porcentagem da população urbana pelo IDH - 2014")

# Para usar as fórmulas da aula passada, precisamos calcular COV(X,Y), VAR(X) e VAR(Y), média de X e média de Y. 
# Neste nosso caso, X é a variável IDH e Y é a variável % da População Urbana .
cov = dados1["IDH"].cov(dados1["% da População Urbana"])

var_X = dados1["IDH"].var()
var_Y = dados1["% da População Urbana"].var()

mX = dados1["IDH"].mean()
mY = dados1["% da População Urbana"].mean()

# Agora podemos calcular a inclinação da reta de ajuste e a constante:
a = cov / var_X
b = mY - a*mX

plt.plot((dados1["IDH"].min(), dados1["IDH"].max()), 
         (a*dados1["IDH"].min() + b, a*dados1["IDH"].max() + b), color='red')
plt.grid(True)
```
